# <font color = 'blue'>Lesson 3: Working with Strings<font>

There are several in-built functionality to manipulate strings. Two are shown below. One which indexes particular character in string variable based on indexing. Another one makes a call to length() function which counts number of character in any string variable.

```python
%%writefile stringmanipulation.cpp
#include <iostream>
using namespace std;

int main()
{

  string name = "Leonhard Euler";
  char first_initial = name[0];
  int name_length =  name.length();  // Here we are using  lenght() functionality to acquire how many characters long is the name. Remember! it also counts empyt space as character.
  cout<< name << " was a great polymath." << endl ;

  cout<<"Initial of "<<name<< " is "<< first_initial <<endl;
  cout<<name<< " is "<< name_length << " characters long name."<< endl;
  return 0;


}
```

# <font color = 'blue'> Lesson 4: Working with numbers<font>

Most of the basic numerical operation (-, +, /, *) are similar. You can directly take numbers and perform these operations. Besides, C++ has a library called 'cmath' that we need to import to perform some in-built math functions such as finding square root, power, max value, etc.

```python
%%writefile lesson4.cpp
#include <iostream>
#include <cmath>     // Importing cmath library
using namespace std;

int main()
{
  int x1 = 20;
  double x2 = 10.35;
  int square_root  = sqrt(81);  // Making use of sqrt() in-built function from cmath library
  cout<< "Addition of "<< x1 << " and " << x2 << " gives " << x1 + x2<<  endl;
  cout<< "Squar root of 81 is "<<square_root<<endl;
 return 0;
}
```
