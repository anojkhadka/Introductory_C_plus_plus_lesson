# <font color = 'blue'>Lesson 2: Function with parameters<font>

We can define a function with parameters which can be used inside to perform certain task.

```python
%%writefile lesson2.cpp
#include <iostream>
using namespace std;

void add_two_numbers(int num1, int num2)

{
cout<<"Sum of "<<num1<< " and " << num2 << " is " << num1 + num2;
}


int main()

{
add_two_numbers(20, 30);  // This is called calling a function. We make a function call by its name and supply with parameters, if any.
return 0;
}
```

```python
!g++ lesson2.cpp -o lesson2
```

```python
!./lesson2
```

<font color = 'magenta'>What happens if we define our function below int main() block? <font>

In that case we need to add the first line of function on top of the in main() block as shown in the example below.

# <font color = 'blue'>Lesson 3: Function defined below int main() block<font>


```python
%%writefile lesson3.cpp
#include <iostream>
using namespace std;

// If we define our function below the 'int main()' block of code, we need to specify the first line of that function above the 'int main()' block so that it gets called
// else it won't.

void fullname(string first, string last);  // Need to specify first line of any function defined below int main() block of code above the 'int main()' block.

int main()

{

fullname("Mintaka", "Particle");

return 0;
}


// Defining the function below the actual 'int main()' block

void fullname(string first, string last)

{
cout<< "Your fullname is "<< first <<" " << last;
}
```

```python
!g++ lesson3.cpp -o lesson3
```

# <font color = 'blue'>Lesson 4: Return function<font>

All the function we defined above are void functions. Now we delve into return function. Return functions are those which not only perform given set of task but also returns some output.

```python
%%writefile lesson4.cpp
#include <iostream>
using namespace std;

double multiplication(double num1, double num2)

{
  double product = num1*num2;
  return product;              // This line tells that when we make a call to this multiplication function it will 'return' whatever value is stored in the product variable.
}

int main()

{
double first_num = 5.5;     //Variable assignments
double second_num = 2.0;

double val = multiplication(first_num, second_num);   // multiplication function is called and this will return the value so that it gets stored in variable 'val'.

cout<< "Product of "<<first_num<< " and "<<second_num<< " is "<< val;

return 0;
}
```

```python
!g++ lesson4.cpp -o lesson4
```

```python
!./lesson4
```
