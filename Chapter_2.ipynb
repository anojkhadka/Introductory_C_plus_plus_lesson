# <font color = 'blue'> Lesson 1: Getting user input<font>

In this lesson we learn how to get input from users. In cases of applications sucha as form fill ups you need to take information from users. This lesson will be helpful to program such scenario.

<font color = 'magenta'>To input single character[link text](https://)<font>

```python
%%writefile lesson1a.cpp
#include <iostream>
using namespace std;

int main()

{

// cin functionality is useful when you just want to input a single character(letter, numbers, etc).
char val;

cout<< "\n Please enter your gender ";
cin>> val;                                // takes user input
cout << "\n Your gender is: "<< val;

return 0;

}
```

```python
!g++ lesson1a.cpp -o lesson1a
```

<font color = 'magenta'>To input words <font>

```python
%%writefile lesson1b.cpp
#include <iostream>
using namespace std;

int main()
{

// However, if you want to input words 'getline()' function is useful as shown below.

string var_name;
cout<< "\n Please enter your name: ";
getline(cin, var_name);
cout<< "\n Your name is "<< var_name;

return 0;

}
```

# <font color = 'blue'> Lesson 2: Arrays <font>
