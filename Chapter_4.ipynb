# <font color = 'blue'>Operators<font>

While writing a program we come across situations where we need to test certain conditions. In order to accmplish this we use certain operators. First, we list out logical operators.

<font color = 'magenta'>Logical Operator<font>

1.   &&     ---> AND  (Is true only when all the condition are satisfied)

2.   ||     ---> OR   (Is true when any of the two conditions is satisfied)

3.   !      ---> NOT  (Used to negate the condtion)

<font color = 'magenta'>Comparison Operator<font>

1.  ==    equal to

2.  !=    not equal to

3.  <     less than

4.  \>     greater than

5.  <=    less than or equal to

6.  \>=    greater than or equal to



# <font color = 'blue'>Lesson 1: If conditional statements<font>


```python
%%writefile lesson1.cpp
#include <iostream>
using namespace std;

int main()

{
bool isweekday = true;  // Defining some boolean condition for test
bool isweekend = true;

// Below is a simple 'if  else if' conditional statement. Condition are strictly specified inside the () brackets and statements to be executed are enclosed inside curly brackets{}.
// Flow of program will be in sequential order i.e. if first condition is satisfied it stops there if not moves to next condition and so on.

if (!isweekday)
{
cout<<"You got to go to work.";
}
else if(!isweekend)

{
cout<<"You can stay home.";
}
else
{
  cout<<"Wait for the decision";
}
return 0;
}
```

# <font color = 'blue'>Lesson 4: Simple calculator with switch case conditional statements<font>

Functionally, 'if else if' conditional statements and switch case statements are similar. However, in some scenarios where we serially switch between some kind of operation 'switch' conditional statements can be more convenient.   

```python
%%writefile lesson4.cpp
#include <iostream>
using namespace std;


void calculator();   // For function defined below int main() block we are required to state the first line of function before int main() block.

int main()
{
  // Calling calculator function
  calculator();
  return 0;
}

// A void calculator function defined

void calculator()
{
  // Declaration of variables
  double first_num, second_num, result;
  int operation_num;

  // Asking for user input
  cout<<"\n Please enter your first number: ";
  cin>>first_num;
  cout<<"\n Please enter your second number: ";
  cin>>second_num;

  cout<<"\n Select what mathematical operation you want to perform from list below using the operation number: \n\n";
  cout<<"1 - Addition \n";
  cout<<"2 - Subtraction \n";
  cout<<"3 - Multiplication \n";
  cout<<"4 - Division \n";
  cout<<" \n\n";
  cin>>operation_num;

// Using switch case conditional statement to switch 'operation_num' variable

switch(operation_num)
{
case 1:
result = first_num + second_num;
cout<<"\n\n Your result on addition is: "<<result;
break;

case 2:
result = first_num - second_num;
cout<<"\n\n Your result on subtraction is: "<<result;
break;

case 3:
result = first_num*second_num;
cout<<"\n\n Your result on multiplication is: "<<result;
break;

case 4:
if (second_num == 0)
{
  cout<<"\n\n Can't divide by zero";
  break;
}
result = first_num/second_num;
cout<<"\n\n Your result on division is: "<<result;
break;

default:  // This section is executed if any of the above condition is not satisfied.
cout<<"\n\n Invalid operation selected";
break;
}

}
```
