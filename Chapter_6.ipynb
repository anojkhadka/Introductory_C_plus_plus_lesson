### <font color = 'blue'>Lesson 1: Intro to Class<font>

Class is important when we do object oriented programming(OOP). C++ follows both object oriented and procedural programming paradigm. In OOP we make a lot of use of classes. Class generally refers to certain group of data that can be bundled in a same category. For instance, we want to keep records of each stduent in a school where each student will have certain type of information such as name, grade, gpa, etc. We can create a class for this and then each student will be an 'object' of this class.




```python
%%writefile lesson1.cpp
#include <iostream>
using namespace std;

// Creating class student records which will hold information declared in three variables: name, grade and gpa.

class student_records{
public:
string name;
int grade;
double gpa;

//CONSTRUCTORS: Below is an example of 'constructors'. We can use the class name above and use it to create constructors. It is like a function where we define all the common
// variables for objects we will create later.
student_records(string arg_name, int arg_grade, double arg_gpa)
{
name = arg_name;
grade = arg_grade;
gpa = arg_gpa;
}

// OBJECT FUNCTION: Below we create a separate function that will perform certain type of task. For example, let's say we want the name of student begin printed out.
// We can just write a function that wiil print the name of student being called. Many such object function can be defined based on the necessity.

// First object function
string name_calling()
{
  cout<<"Your name is "<<name;
  return name;
}

// Second object function
int which_school()
{
  if(grade<=5){
    cout<<"\n\nYou are in primary";
  }
else if(grade>5 && grade<=10)
{
  cout<<"\n\nYour are in middle school.";
}
else{
  cout<<"\n\nYour are in high school";
}
return grade;
}

};



int main()

{

// Once we create class we can then instantiate objects for the class. Each object will hold data defined in class above. The advantage of creating class to write OOP is that
// we don't have to separately specify name, grade gpa variable for each student again and again. Once definded in the class above, we can then access them for all objects.
// Below are three objects we created for 'student_records' class we created above.

student_records student1("Nasim", 10, 3.4);

student_records student2("George", 12, 3.8);

student_records student3("Tandav", 5, 3.0);


student2.name_calling(); // This will call the 'name_calling' function which is a class function.
student2.which_school();

return 0;
}

```

###<font color = 'blue'>Lesson 3: Inheritance in class<font>

This tells us at the very basic level how we can import(or inherit) functionalities from one class to another class. This is very useful in object oriented programming when we need to use same function in different classes. We don't need to write the function again and again in all the classes that use it. Instead we can inherit it from just one place to whereever it is required. It's not only the function that we can inherit. We can also inherit variables and class itself. This helps us to keep our code clean and concise by allowing us to inherit functions(or varialbes) that are used in multiple classes. This code below just shows one very simple inheritance example of one function being used in two different classes. There are lot more things to learn about inheritance.

```python

%%writefile lesson3.cpp
#include <iostream>
using namespace std;

// Creating class student records which will hold information declared in three variables: name, grade and gpa.

class student_records{
public:

void identity()
{
  cout<<"\n\n Your are good person "<<endl;
}

};

// Below is a separate class different from the class above. For instance above class has an 'identity' function which prints something. We want to do the same thing in our new
// class as well. We don't need to write the function again inside this class. Instead we can import/inherit the function from the previous class. In the initial lines we see
// we have inherited 'pulic student_records' which is our previous class. This will allow us to import the functionality of previous class into our new class.

class employee_record :
public student_records{
public:

void welcome_message()
{
  cout<<"\n\n You are welcome to the office "<<endl;
}
};


int main()

{
employee_record employee1;
employee1.identity();            // 'identity' function from previous class is inherited to this class employee record.
employee1.welcome_message();

return 0;
}
```
